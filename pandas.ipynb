## 4.Manipulating data

```python
missing_data = pd.read_csv("./assets/data/393 car-sales-missing-data.csv")
missing_data.head()
```

```python
# Fill NA 
# inplace=True returns None và thay đổi trực tiếp trên df
# inplace=False (default) returns bản copy data và thay đổi trên df_copy
missing_data.Odometer.fillna(missing_data.Odometer.mean())
```

```python
# Drop NA 
missing_data.dropna()
```

```python
# Convert price obj to number
df = pd.DataFrame({"price": ['$40.69', '$1,100.00', '$12.34']})
def convert_price(df,column):
    df[column] = df[column].astype(str).str.replace('[$,]', '',regex =True).astype('float')
    df[column]= df[column].round(0) # làm tròn số thập phân
    df[column] = df[column].astype('int64') # chuyển sang int 
convert_price(df,'price')
df
```

```python
# Drop column
# convert_price(data,'Price')
data.drop(columns='Doors')
missing_data
```

```python
# Find NAN
missing_data.loc[missing_data["Price"].isna(),"Price"]

```

```python
# Where (cond , orther = nan ) : chọn ra những giá trị điều kiện true 
#  cond - điều kiện , orther : thay thế những giá trị false 
missing_data.where( missing_data["Price"].isna(),2)
```

```python
# mask ngược lại với where
missing_data.mask( missing_data["Price"].isna(),2)
```

```python
# apply : giống ham map()
missing_data.Doors.apply(lambda x : x*x)
```

```python
# rename column
missing_data.rename(columns={
    'Doors': 'New Doors'
})
```
