## 1.Datatypes and attributes 

```python
# Numpy`s main datatype is ndarray 
# - 1-Dimensional : vector
# - More than 1-Dimensional : matrix  

a1 = np.array([1,2,3])
a3 = np.array([
    [[1, 2, 3],
     [4, 5, 6]],

    [[7.5, 8.0, 9],
     [10, 11.5, 12]]
])
print ('>>> Shape :' , f'a1: {a1.shape} | a3: {a3.shape} (h,r,c)')

```

```python
print ('>>> Dim :' , f'a1: {a1.ndim} | a3: {a3.ndim}')
```

```python
print ('>>> Datatype :' , f'a1: {a1.dtype} | a3: {a3.dtype}')
print ('>>> Type : ', f'a1: {type(a1)} | a3: {type(a3)}')
```

```python
print ('>>> Check number elements :' , f'a1: {a1.size} | a3: {a3.size}')
```

## 4. Reshape and Transpose

```python
# Broadcasting là một cơ chế mạnh mẽ cho phép thực thi các phép toán số học trên các numpy array có kích thước khác nhau.
# A + B , A* B ,....
# Rule :
# Hai dimensions  tương thích khi
# + chúng bằng nhau, hoặc
# + một trong số họ là 1
# VD :
# A      (4d array):  8 x 1 x 6 x 1
# B      (3d array):      7 x 1 x 5
# Result (4d array):  8 x 7 x 6 x 5
# Lưu ý : nếu mảng không tương thích thì có thể dùng reshape để cho hai array tương thích 
# 2. Transpose : ma trẩn chuyển vị 

a3_T = a3.T
print (f'>>>> a3 : {a3.shape} | Transpose : {a3_T.shape}' )
```
